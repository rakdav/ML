```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
```

```python
df = pd.read_csv(url)
df
```

```python
df=df[['Survived','Pclass','Age','SibSp','Parch']]
```

```python
df.dropna(inplace=True)
df
```

```python
X=df[['Pclass','Age','SibSp','Parch']]
y=df['Survived']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_scaled
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
log_model=LogisticRegression()
```

```python
log_model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_pred = log_model.predict(X_test)
```

```python
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Точность модели: {accuracy:.2f}")
print(classification_report(y_test, y_pred))
```

```python
feature_names = ['Pclass', 'Age', 'SibSp', 'Parch']
coefficients = log_model.coef_[0]
# Визуализируем влияние факторов
plt.figure(figsize=(8, 5))
sns.barplot(x=feature_names, y=coefficients)
plt.xlabel("Признаки")
plt.ylabel("Коэффициенты модели")
plt.title("Влияние факторов на вероятность выживания")
plt.show()
```

```python
Arina = {
    'Pclass': [1],
    'Sex': ['female'],
    'Age': [38],
    'SibSp': [1],
    'Parch': [0],
    'Fare': [8.25],
    'Embarked': ['S']
}
df_new = pd.DataFrame(Arina)
df_new['Sex'] = df_new['Sex'].map({'male': 0, 'female': 1})
df_new['Embarked'] = df_new['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
df_new['Age'].fillna(28, inplace=True)
df_new['Fare'].fillna(14.45, inplace=True)
df_new
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_new)
X_scaled
```

```python
prediction_proba = log_model.predict_proba(X_scaled)
print(f"Вероятность выживания: {prediction_proba[0][1]:.2f}")
class_prediction = log_model.predict(X_scaled)
print(f"Предсказанный класс: {class_prediction[0]}")
```
